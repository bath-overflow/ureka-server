```python
elements_no_header_footer = [
    el for el in elements if el.category not in ["Header", "Footer"]
]
```

```python
from pprint import pprint

for el in elements_no_header_footer:
    pprint(el.to_dict())
```

```python
for el in elements_no_header_footer:
    print("{:10} {:10} {:40}".format(el._element_id[:6], el.metadata.parent_id[:6] if el.metadata.parent_id is not None else "", el.text))
```

```python
from unstructured.chunking.basic import chunk_elements

chunks = chunk_elements(elements_no_header_footer)
```

```python
# View the original elements and their metadata
chunk = chunks[0]
for el in chunk.metadata.orig_elements:
    pprint(el.to_dict())
```
